```python
## import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##loading boston dataset
from sklearn.datasets import load_boston
boston=load_boston()
```

```python
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df.head()
```

```python
#to find the correlation we use corr()
boston_df.corr()
```

```python
# visualizing the correlation
plt.figure(figsize=(14,14))
sns.heatmap(boston_df.corr(),annot=True,cmap='YlGnBu')
#annot=True display the correlation values
```

```python
# correlation greater than 0.7
correlation=boston_df.corr()
cor=correlation[correlation>0.7]
cor
plt.figure(figsize=(13,13))

sns.heatmap(cor,cmap='Greens',annot=True)
```

```python
#checking the correlation between 0.7 and 0.9
#cor=[(boston_df.corr()>0.7) & (boston_df.corr()<0.9)]-->gives boolean values
cor=correlation[(correlation>0.7)&(correlation<0.9)]
cor
```

```python
## all the columns as input
#drop the columns which are highly correlated and build a MLR;you need to compare the performance
```

```python
boston_df['MEDV']=boston.target
boston_df.head()
```

```python
## split the data into training and testing ;build the model on training data and test the model on testing data
```

```python
## split the data into input and output
x=boston_df.drop(['MEDV'],axis=1)
y=boston_df.MEDV#bosto_df['MEDV']
```

```python
# divide the x and y into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#build the model on training data-->x_train and y_train
# sklearn library:import ,instantiate,fit
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
# checking  intercept snd coefficient values
print(lm.intercept_)
lm.coef_
```

```python
## create a dataframe for coefficients
coefficients=pd.DataFrame([x_train.columns,lm.coef_]).T
coefficients
# .T helps u to get the values in a column
```

```python
#multi regression
#y=b0+b1x1+b2x2+....b13x13 as we have 13 columns
#y=36.93325545711978+(-0.117735)()
```

```python
## checking the model prediction on training data
y_train_pred=lm.predict(x_train)
y_train_pred
```

```python
## we need to compare the actual values(y_train) and the predicted values (y_train_pred)
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('R^2:',r2_score(y_train,y_train_pred))
print('Adjusted R^2:',1- (1-r2_score(y_train,y_train_pred))*(len(x_train)-1)/(len(x_train)-x_train.shape[1]-1))

print('MAE:',mean_absolute_error(y_train,y_train_pred))
print('MSE:',mean_squared_error(y_train,y_train_pred))
print('RMSE:',np.sqrt(mean_squared_error(y_train,y_train_pred)))
```

```python
## regplot(y_train-->actual values and y_train_pred-->predicted values)
sns.regplot(y_train,y_train_pred)
```

```python
# visualize the differences between actual values and predicted values
plt.scatter(y_train,y_train_pred)
plt.xlabel('Price')
plt.ylabel('Model Predicted Price')
```

```python
#y_train-y_train_pred==Errors
plt.scatter(y_train_pred,y_train-y_train_pred)
plt.xlabel("Predicted values")
plt.ylabel('Residuals')
```

```python
# test the model on testing data
y_test_pred=lm.predict(x_test)#test data-->unseen data
y_test_pred
```

```python
## we need to compare the actual values(y_test) and the predicted values (y_test_pred)
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('R^2:',r2_score(y_test,y_test_pred))
print('Adjusted R^2:',1- (1-r2_score(y_test,y_test_pred))*(len(x_test)-1)/(len(x_test)-x_test.shape[1]-1))

print('MAE:',mean_absolute_error(y_test,y_test_pred))
print('MSE:',mean_squared_error(y_test,y_test_pred))
print('RMSE:',np.sqrt(mean_squared_error(y_test,y_test_pred)))
```
