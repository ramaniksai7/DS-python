```python
import pandas as pd
import numpy as np
df = pd.read_csv("Match (1).csv")
df
```

# 1.Print ManOfMach count of each player in Hyderabad stadium

```python
city = df.loc[df.City_Name.eq("Hyderabad")]
players = city.groupby(['ManOfMach']) 

manOfmatch = players['ManOfMach'].count()

data = {'no of man of matches':manOfmatch} 
df1 = pd.DataFrame(data)
df1
```

# 2.Plot a pie chart for the country_name column

```python
import matplotlib.pyplot as plt
country_data = df.Country_Name.unique()
sizes = df.groupby(['Country_Name']).count()
colors = ['blue', 'green', 'red']
explode = (0.1, 0, 0)  
plt.pie(sizes.Match_SK,explode=explode, labels=country_data, colors=colors, autopct='%1.1f%%', shadow=True, startangle=45)

plt.axis('equal')
plt.show()
```
